# IGNET use with simulated data

```python
# After having generated the distance matrix of your data, get the output.
%matplotlib inline
import gzip
import cPickle as pkl
import seaborn as sns

from scipy.cluster.hierarchy import fcluster, linkage, dendrogram
from scipy.spatial.distance import squareform
```

```python
with gzip.open('/home/fede/ignet_tests/results/ignet_debug_2016-07-08_12.28.54/ignet_debug_2016-07-08_12.28.54_similarity_matrix.pkl.tz', 'r') as f:
    sm = pkl.load(f)
dm = 1. - sm.toarray()
```

```python
threshold = .15
#dm[dm > threshold] = 1
links = linkage(squareform(dm), method='single')
_ = dendrogram(
    links,
    leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=8.,  # font size for the x axis labels
)
```

```python
clusters = fcluster(links, .15, criterion='distance')
print max(clusters)
```
